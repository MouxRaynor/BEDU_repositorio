```python
#funciones vectorizadas
db_siniestros_mar= db_siniestros_mar.drop(columns=['Id_Embarcacion'])
db_siniestros_mar_mapping={'MONEDA':'Moneda',
                           'FORMA_DE_VENTA':'Forma_Venta',
                           'Tipo Embarcacion ':'Timpo_Embarcacion',
                           'Id_Uso':'Uso',
                           'AÑO DE CONSTRUCCION ':'Anio_Construccion',
                           'Id_Cobertura':'Cobertura',
                           'CAUSA_DEL_SINIESTRO':'Cause',
                           'NUMERO_DE_SINIESTROS':'Num_Siniestros',
                           'MONTO_DEL_SINIESTRO':'Incurrido',
                           'GASTOS_DE_AJUSTE':'Gastos_Ajuste',
                           'SALVAMENTO':'Salvamento',
                           'MONTO_PAGADO':'Monto_Pagado',
                           'MONTO_DE_DEDUCIBLE':'Monto_Deducible',
                           'MONTO_DE_COASEGURO':'Monto_Coaseguro',
                           }
db_siniestros_mar = db_siniestros_mar.rename(columns=db_siniestros_mar_mapping)
db_siniestros_mar.head(5)
```

```python
###Renombrar la moneda
def nombrar_moneda(moneda):
  if moneda==1:
    moneda ='pesos'
  elif moneda==2:
    moneda='dolares'
  return moneda
db_siniestros_mar['Moneda']=db_siniestros_mar['Moneda'].apply(nombrar_moneda)
db_siniestros_mar.head(5)
    
```

```python
#Obtenemos el incurrido real en pesos y quitamos columnas que no nos sirven
import numpy as np
db_siniestros_mar['Tipo_Cambio']=np.where(db_siniestros_mar['Moneda']==1,1,20)
db_siniestros_mar['Incurrido_Homologado']=db_siniestros_mar['Tipo_Cambio']*db_siniestros_mar['Incurrido']
db_siniestros_mar= db_siniestros_mar.drop(['Gastos_Ajuste'], axis=1)
db_siniestros_mar= db_siniestros_mar.drop(['Salvamento'], axis=1)
db_siniestros_mar=db_siniestros_mar.drop(['Monto_Pagado'], axis=1)
db_siniestros_mar=db_siniestros_mar.drop(['Monto_Deducible'], axis=1)
db_siniestros_mar=db_siniestros_mar.drop(['Monto_Coaseguro'], axis=1)
columnas=list(db_siniestros_mar.columns)
```
